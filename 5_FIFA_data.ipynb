```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
fifa_data = pd.read_csv('fifa_data.csv')
fifa_data.head(6).sort_values(['Overall','Name'],ascending = [0,0])
```

```python
plt.figure(figsize=(10,5),dpi=360)

plt.hist(fifa_data['Overall'],np.arange(40,101,10),color = '#24F290')

plt.xticks(np.arange(40,101,10),color = 'white')
plt.yticks(color = 'white')

plt.title('Distribution of players skilles in FFIFA 2019',fontdict = {'fontname':'Arial','fontsize':'20','fontweight':'bold'},color = 'white')
plt.xlabel('Skil level',fontdict = {'fontname':'Arial','fontsize':'20'},color = 'white')
plt.ylabel('Numper of players',fontdict = {'fontname':'Arial','fontsize':'20',},color = 'white')

plt.show()
```

# Bie chart

```python

Left = fifa_data.loc[fifa_data['Preferred Foot'] == 'Left'].count()['ID']
Right = fifa_data.loc[fifa_data['Preferred Foot'] == 'Right'].count()['ID']

plt.figure(dpi = 200)

plt.pie([Left,Right], autopct ='%.2f %%' ,labels = ['Left','Right'],colors = ['#6EC7F5','b'])

plt.title('Preferred Foot of FIFA players',color = 'w')
#plt.legend()
plt.show()
```
