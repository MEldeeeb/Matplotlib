# Gas prices over time 

```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
gas_data = pd.read_csv('gas_prices.csv')
gas_data
```

```python
plt.figure(figsize = (15,7),dpi = 300)

# ploting all gas prices for each country vs years
for country in gas_data.head():
    if country != 'Year':
        plt.plot(gas_data['Year'],gas_data[country],label = country ,linestyle = '--',linewidth = 1,marker = '.',markersize = 5)
        
# formating title and labels
plt.title('Gas prices over time', fontdict = {'fontname' :'comic sans ms','fontweight':'bold','fontsize':30},color = 'white')
plt.xlabel('year', fontdict = {'fontname' :'comic sans ms','fontsize':30},color = 'white')
plt.ylabel('price', fontdict = {'fontname' :'comic sans ms','fontsize':30},color = 'white')

# creating legend
plt.legend()

# adjusting the ticks 
plt.xticks(gas_data['Year'][:19:2],color = 'white')
plt.yticks(np.arange(0,9,1),color = 'white')

#saving the figuer
plt.savefig('Gas_price.png',dpi=720)

plt.show()
```
